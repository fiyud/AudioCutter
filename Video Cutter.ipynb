```python
import os
import librosa
from pydub import AudioSegment
```

```python
def cut_audio_into_segments(audio, segment_duration):
    segments = []
    total_duration = len(audio)
    segment_start = 0
    segment_end = segment_duration

    while segment_end <= total_duration:
        segment = audio[segment_start:segment_end]
        segments.append(segment)
        segment_start += segment_duration
        segment_end += segment_duration

    #add extra if it's less than 2s (or you can change it to your will)
    if segment_start < total_duration:
        segment = audio[segment_start:]
        segments.append(segment)

    return segments

def process_audio_folder(input_folder, output_folder):
    for filename in os.listdir(input_folder):
        file_path = os.path.join(input_folder, filename)
        #choose your file format here, here i am dealing with wav files so leave the format as "wav"
        if os.path.isfile(file_path) and filename.endswith(".wav"):
            audio = AudioSegment.from_file(file_path)

            #pydub will count on milisecon
            segment_duration = 2000

            #Cut into segments
            segments = cut_audio_into_segments(audio, segment_duration)

            #Export audio to many files
            for i, segment in enumerate(segments):
                output_file_path = os.path.join(output_folder, f"segment_{i+1}_{filename}")
                segment.export(output_file_path, format="wav")
                print(f"Done: {output_file_path}")

input_folder = "/content/drive/MyDrive/"#your_inputfolder_path
output_folder = "/content/drive/MyDrive/"#your_outputfolder_path

process_audio_folder(input_folder, output_folder)
```
